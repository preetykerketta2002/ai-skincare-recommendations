```python
import os
import pandas as pd

dataset_path = "dataset"

rows = []

for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    
    if not os.path.isdir(folder_path):
        continue
    
    for file in os.listdir(folder_path):
        row = {
            "filename": file,
            "acne_inflammatory": 0,
            "acne_noninflammatory": 0,
            "hyperpigmentation": 0,
            "wrinkles": 0
        }
        
        # mark correct class
        if folder == "acne_inflammatory":
            row["acne_inflammatory"] = 1
        elif folder == "acne_noninflammatory":
            row["acne_noninflammatory"] = 1
        elif folder == "hyperpigmentation":
            row["hyperpigmentation"] = 1
        elif folder == "wrinkles":
            row["wrinkles"] = 1
        
        rows.append(row)

df = pd.DataFrame(rows)
df.to_csv("labels.csv", index=False)

print("DONE ✅ labels.csv created")
```

```python
import os
import pandas as pd

dataset_path = "dataset"

rows = []

for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    
    if not os.path.isdir(folder_path):
        continue
    
    for file in os.listdir(folder_path):
        row = {
            "filename": f"{folder}/{file}",  # 🔥 THIS IS THE CHANGE
            "acne_inflammatory": 0,
            "acne_noninflammatory": 0,
            "hyperpigmentation": 0,
            "wrinkles": 0
        }
        
        row[folder] = 1
        rows.append(row)

df = pd.DataFrame(rows)
df.to_csv("labels.csv", index=False)

print("CSV updated ✅")
```

```python
import pandas as pd
import numpy as np
import os
from tensorflow.keras.preprocessing.image import load_img, img_to_array

df = pd.read_csv("labels.csv")

images = []
labels = []

for i in range(len(df)):
    img_path = os.path.join("dataset", df.iloc[i]['filename'])
    
    img = load_img(img_path, target_size=(224,224))
    img = img_to_array(img) / 255.0
    
    label = df.iloc[i][1:].values.astype(float)
    
    images.append(img)
    labels.append(label)

X = np.array(images)
y = np.array(labels)
```
